```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Corelation matrix calculation
correlation_matrix = dataset.corr()

# Visualization of corelation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlaction matrix between features")
plt.show()

# Check corelation features
print("\nFeature corelation 'Class':")
print(correlation_matrix["Class"].sort_values(ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
import numpy as np

# Load dataset
data = pd.read_csv("creditcard_dataset.csv")

# Drop non-feature columns
X = data.drop(columns=["Class"])
y = data["Class"]

# Reduce dataset size if needed
sample_size = min(10000, len(data))  # Limit dataset to 10,000 samples if it's too large
X_sampled, y_sampled = X.sample(sample_size, random_state=42), y.sample(sample_size, random_state=42)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_sampled)

# Apply t-SNE with optimized settings
tsne = TSNE(n_components=2, perplexity=30, n_iter=500, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# Apply UMAP with optimized settings
umap_model = umap.UMAP(n_components=2, n_neighbors=15, random_state=42)
X_umap = umap_model.fit_transform(X_pca)

# Visualize t-SNE results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y_sampled, palette={0: "blue", 1: "red"}, alpha=0.7)
plt.title("t-SNE Fraud Detection Visualization")
plt.show()

# Visualize UMAP results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y_sampled, palette={0: "blue", 1: "red"}, alpha=0.7)
plt.title("UMAP Fraud Detection Visualization")
plt.show()
```
