```python
# Implementation XGBoost ML algorythm to check fraud transaction
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import xgboost as xgb
```

```python
# Load data from CSV file
dataset = pd.read_csv("creditcard_dataset.csv")
```

```python
# Podział na cechy (X) i etykiety (y)
X = dataset.drop(columns=["Class"])  # Wszystkie kolumny oprócz "Class"
y = dataset["Class"]  # Kolumna docelowa
```

```python
# Podział na zbiór treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Initialize XGBoost Classifier
model = xgb.XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, subsample=0.8, colsample_bytree=0.8, random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predykcja na zbiorze testowym
y_pred = model.predict(X_test)
```

```python
# Ocena modelu
accuracy = accuracy_score(y_test, y_pred)
print(f"Dokładność modelu: {accuracy:.4f}")
```

```python
# Wyświetlenie klasyfikacji i macierzy błędu
print("\nRaport klasyfikacji:")
print(classification_report(y_test, y_pred))
```

```python
print("\nMacierz konfuzji:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Calculate percentage of each class in training and validation sets
train_class_distribution = y_train.value_counts(normalize=True) * 100
test_class_distribution = y_test.value_counts(normalize=True) * 100

# Print the percentages
print("Class distribution in Training Set:")
print(train_class_distribution)

print("\nClass distribution in Validation Set:")
print(test_class_distribution)
```
